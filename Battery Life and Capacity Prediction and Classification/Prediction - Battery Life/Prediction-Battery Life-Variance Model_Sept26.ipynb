# train test

```python
# X = data['var_dis_cap'].values.reshape(-1,1)
# y = data['cycle_life'].values.reshape(-1,1)

X_train, X_test, y_train, y_test = train_test_split(x5.values.reshape(-1,1), y5.values.reshape(-1,1), test_size=0.2, random_state=42)
```

# Linear Regression

```python
# model fitting 
regressor = LinearRegression()

#training the algorithm
regressor.fit(X_train, y_train) 

print( regressor.intercept_)
print( regressor.coef_)
```

```python
# X1_test = pd.DataFrame(df['var_dis_cap'])
# y1_act = pd.DataFrame(df['cycle_life'])

y5_pred_reg= regressor.predict(X_test)
```

```python
plt.figure(figsize=(20,10))
plt.scatter(X_test, y_test,  color='gray')
plt.plot(X_test, y5_pred_reg, color='red', linewidth=6)

plt.grid(b=True, which='major', color='#666666', linestyle='-')

plt.show()
```

# SVR

```python
from sklearn.svm import SVR

clf = SVR(gamma='scale', C=1.0, epsilon=0.2)
clf.fit(X_train, y_train) 

# Prediction
y1_pred_svr = clf.predict(X_test)

# df = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred.flatten()})
# df.display()

plt.figure(figsize=(20,10))
plt.scatter(X_test, y_test,  color='gray')
plt.plot(X_test, y1_pred_svr, color='red', linewidth=6)

plt.grid(b=True, which='major', color='#666666', linestyle='-')

plt.show()
```

```python
print('Mean Absolute Error:', mean_absolute_error(y_test, y5_pred_reg))  
print('Mean Squared Error:', mean_squared_error(y_test, y5_pred_reg))  
print('Root Mean Squared Error:', np.sqrt(mean_squared_error(y_test, y5_pred_reg)))
```
